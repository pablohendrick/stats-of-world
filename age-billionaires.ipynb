```python
#importing libraries...
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Billionaires Statistics Dataset.csv')
```

```python
data.head(10)
```

```python
sns.set(style="ticks", palette="Set2")
```

```python
plt.figure(figsize=(12, 6))

sns.pairplot(data[['age']], height=5, aspect=2)
plt.suptitle('Pairplot - age', y=1)
plt.tight_layout()
plt.show()
```
